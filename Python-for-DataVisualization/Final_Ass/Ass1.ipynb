```python
import pandas as pd
```

```python
df = pd.read_csv("https://cocl.us/datascience_survey_data",index_col=0)
```

```python
df.index.name=None
df
```

```python
import matplotlib.pyplot as plt
```

```python
df.sort_values(by="Very interested",ascending=False,inplace=True)
```

```python
df_pc = df/2233
df_pc.head()
```

```python
fig = plt.figure()
ax = df_pc.plot(kind="bar",
           figsize=(20,8),
           width=0.8,
           color=["#5cb85c","#5bc0de","#d9534f"]
          )
for i,spine in enumerate(plt.gca().spines.values()):
    if i != 2 :
        spine.set_visible(False)
plt.legend(fontsize=14)
plt.tick_params(top='off', 
                bottom='off', 
                left='off', 
                right='off', 
                labelleft='off', 
                labelbottom='on',
                labelsize=14  
               )
plt.title("Percentage of Respondents' Interest in Data Science Areas",fontsize=16)
pdict={0:-0.37,1:-0.1,2:0.17}
for i in range(0,len(df)):
    for j in range(0,len(df.iloc[0,:])):
        value = df_pc.iloc[i,j]
        label = "{:.2f}%".format(value*100)
        plt.annotate(label,
                     xy=(i+pdict[j],value+0.01))
plt.show()
```
